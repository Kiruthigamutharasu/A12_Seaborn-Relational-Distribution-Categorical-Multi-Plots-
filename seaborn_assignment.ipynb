Dataset Name: Students Performance in Exams
Kaggle Link: https://www.kaggle.com/datasets/spscientist/students-performance-in-exams


```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("StudentsPerformance.csv")

# View first 5 rows
df.head()
```

TASK 1 â€“ RELATIONAL PLOT

```python
# Task 1.1 - Relational Plot
#2. Use categorical column (gender as hue)
sns.relplot(data=df,
            x="math score",
            y="reading score",
            hue="gender")
plt.title("Math Score vs Reading Score")
plt.show()
```

```python
# Task 1.3 - Scatter Style
sns.relplot(data=df,
            x="math score",
            y="reading score",
            style="gender")
plt.title("Scatter Style Relplot")
plt.show()
```

TASK 2 â€“ LINE & FACET

```python
#1. Line plot
sns.lineplot(data=df,
             x="math score",
             y="writing score")
plt.title("Line Plot: Math vs Writing")
plt.show()
```

```python
#2. Convert to scatter-style
sns.scatterplot(data=df,
                x="math score",
                y="writing score")
plt.title("Scatter Version")
plt.show()
```

```python
#3. Faceting using categorical column
sns.relplot(data=df,
            x="math score",
            y="writing score",
            col="gender",
            kind="scatter")
plt.show()
```

TASK 3 â€“ DISTRIBUTION

```python
#1. Histogram
sns.histplot(df["math score"])
plt.title("Histogram of Math Score")
plt.show()
```

```python
#2. KDE Plot
sns.kdeplot(df["math score"])
plt.title("KDE Plot")
plt.show()
```

```python
#3. Rug Plot
sns.rugplot(df["math score"])
plt.title("Rug Plot")
plt.show()
```

```python
#4. Histogram + KDE
sns.histplot(df["math score"], kde=True)
plt.title("Histogram + KDE")
plt.show()
```

TASK 4 â€“ BIVARIATE

```python
#1. Bivariate Histogram
sns.histplot(data=df,
             x="math score",
             y="reading score")
plt.title("Bivariate Histogram")
plt.show()
```

```python
#2. Bivariate KDE
sns.kdeplot(data=df,
            x="math score",
            y="reading score")
plt.title("Bivariate KDE")
plt.show()
```

TASK 5 â€“ MATRIX PLOTS

```python
#1. Pairplot
sns.pairplot(df[["math score","reading score","writing score"]])
plt.show()
```

```python
#2. Heatmap Correlation
corr = df[["math score","reading score","writing score"]].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()

```

TASK 6 â€“ CATEGORICAL

```python
#1. Bar Plot
sns.barplot(data=df,
            x="gender",
            y="math score")
plt.title("Bar Plot")
plt.show()
```

```python
#2. Box Plot
sns.boxplot(data=df,
            x="gender",
            y="math score")
plt.title("Box Plot")
plt.show()
```

```python
#3. Violin Plot
sns.violinplot(data=df,
               x="gender",
               y="math score")
plt.title("Violin Plot")
plt.show()
```

```python
#4. Count Plot
sns.countplot(data=df,
              x="gender")
plt.title("Count Plot")
plt.show()
```

TASK 7 â€“ REGRESSION

```python
#1. Regression Plot
sns.regplot(data=df,
            x="math score",
            y="reading score")
plt.title("Regression Plot")
plt.show()
```

```python
#2. Lmplot with categorical
sns.lmplot(data=df,
           x="math score",
           y="reading score",
           hue="gender")
plt.show()
```

TASK 8 â€“ MULTI PLOTS

```python
#1. FacetGrid
g = sns.FacetGrid(df, col="gender")
g.map(sns.scatterplot, "math score", "reading score")
plt.show()
```

```python
#2. Multi-plot Dashboard
sns.relplot(data=df,
            x="math score",
            y="reading score",
            col="gender")
plt.show()

sns.catplot(data=df,
            x="gender",
            y="math score",
            kind="box")
plt.show()

sns.displot(df["math score"])
plt.show()
```
